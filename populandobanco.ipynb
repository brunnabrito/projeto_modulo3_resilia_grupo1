## Lendo o CSV com os municípios do Brasil e Estados

```python
cidades = pd.read_csv("/Users/brunnabrito/Documents/Data_Analitycs/Resilia/Projetos finais/projeto_modulo3_resilia_grupo1/Municipios-Brasileiros-main/csv/municipios.csv",
                        usecols=[1,5])
cidades.head()
```

```python
siglas_estados = pd.read_csv("/Users/brunnabrito/Documents/Data_Analitycs/Resilia/Projetos finais/projeto_modulo3_resilia_grupo1/Municipios-Brasileiros-main/csv/estados.csv",
                     usecols=[0,1])
siglas_estados.head()
```

```python
municipios = pd.merge(cidades, siglas_estados, left_on="codigo_uf", right_on="codigo_uf")
municipios.head()
```

## Escolhendo aleatóriamente os municípios

```python
for i in range(40,200):
    random = randint(0,5569)
    municipio = municipios.loc[random, "nome"]
    estado = municipios.loc[random, "uf"]
    tabela.loc[i, "cidade"] = municipio
    tabela.loc[i, "estado"] = estado
```

```python
display(tabela)
```

## Definindo aleatóriamente o curso

```python
cursos = ["1 - Web DEV","2 - Web DEV","3 - Web DEV","4 - Web DEV","9 - Web DEV","6 - Web DEV","7 - Web DEV","8 - Web DEV", "5 - Web DEV","10 - Web DEV","11 - Web DEV", "12 - Dados"]
for i in range(0,200):
    curso = random.choice(cursos)
    tabela.loc[i, "curso"] = curso
```
